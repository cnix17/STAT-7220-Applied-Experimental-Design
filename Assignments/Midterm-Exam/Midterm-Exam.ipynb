```python
#Question 6

#Exploratory Analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

#load the data
df = pd.read_excel("Video Game Controllers.xlsx")

print(df.head())


#Summary stats by material

summary = df.groupby("Material")["Comfort"].describe()
print("Summary Stats by Material:")
print(summary)



#Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Material", y="Comfort", palette="Set2")
sns.swarmplot(data=df, x="Material", y="Comfort", color="black", alpha=0.5)
plt.title("Comfort Ratings by Material")
plt.xlabel("Material")
plt.ylabel("Comfort Rating")
plt.show()


#Histogram for each group
materials = df["Material"].unique()
fig, axes = plt.subplots(2, 2, figsize=(10,7))
for mat, ax in zip(materials, axes.flatten()):
    sns.histplot(df[df["Material"]==mat]["Comfort"], kde=True, ax=ax)
    ax.set_title(f"{mat} Comfort Distribution")
plt.tight_layout()
plt.show()

#Preliminary one-way ANOVA for reference
f_stat, p_val = stats.f_oneway(
    *[df[df["Material"]==mat]["Comfort"] for mat in df["Material"].unique()]
)
print(f"Preliminary ANOVA F={f_stat:.3f}, p={p_val:.4f}")






#Interpretation
#Summary table: We are looking at means & standard deviation
#Boxplot: Shows us the median and spread by group
#Histogram: Checking the shape of distributions
#p-value preview: if p < 0.05, this will hint that differences may exist

#Our exploratory analysis shows clear differences in mean comfort rating between materials, with silicone having the highest avg comfort and metal the lowest.
#Visualizations show reasonably normal distributions per group
# The preliminary ANOVA yields a p-value <0.001, providing strong evidence against the null hypothesis. this suggests that at least one material type differs significantly in perceived comfort
```

```python
#Question 7

#Normality assumption
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import shapiro
import matplotlib.pyplot as plt
import seaborn as sns

# Build one-way ANOVA model
model = ols('Comfort ~ C(Material)', data=df).fit()

# Get residuals
residuals = model.resid

#Shapiro-Wilk test
shapiro_stat, shapiro_p = shapiro(residuals)
print(f"Shapiro-Wilk Test: Statistic={shapiro_stat:.4f}, p={shapiro_p:.4f}")

#Q-Q Plot
sm.qqplot(residuals, line ='s')
plt.title("Q-Q Plot of Residuals")
plt.show()

#Histogram of residuals
sns.histplot(residuals, kde=True)
plt.title("Histogram of Residuals")
plt.show()



#Interpretation
#The Shapiro-Wilk test returned a p-value of 0.9509, which is greater than 0.05. In turn, we fail to reject the null hypothesis of normality.
#Both the Q-Q plot and histogram of residuals indicate that the data are approximately normally distributed
# The assumption of normality is met for the ANOVA model
```

```python
#Question 8

from scipy.stats import levene
import matplotlib.pyplot as plt

# --- Levene's Test for Homogeneity of Variance ---
groups = [df[df["Material"] == mat]["Comfort"] for mat in df["Material"].unique()]
lev_stat, lev_p = levene(*groups)
print(f"Levene's Test: Statistic={lev_stat:.4f}, p-value={lev_p:.4f}")

# --- Residuals vs. Fitted Plot ---
fitted = model.fittedvalues
residuals = model.resid

plt.scatter(fitted, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.title("Residuals vs. Fitted Values")
plt.xlabel("Fitted values")
plt.ylabel("Residuals")
plt.show()



#Interpretation
#Levene‚Äôs test yielded a p-value of 0.2679, which is greater than 0.05. Therefore, we fail to reject the null hypothesis that the variances across groups are equal.
# Additionally, the residual vs. fitted plot shows a random scatter of points around zero without any obvious funnel or pattern.
# These results indicate that the assumption of homogeneity of variance is met / satisfied
```

```python
#Question 9

from statsmodels.stats.anova import anova_lm

# Run ANOVA table
anova_table = anova_lm(model, typ=2)
print(anova_table)

#Interpretation
#The one-way ANOVA revealed a statistically significant effect of grip material on comfort ratings, F(3, 36) = 47.48, p < 0.001.
# This extremely small p-value provides evidence against the null hypothesis, indicating that at least one material‚Äôs mean comfort rating is significantly different from the others
# Our decision is to reject the null hypothesis
```

```python
#Question 10
import matplotlib.pyplot as plt
from statsmodels.stats.multicomp import pairwise_tukeyhsd, MultiComparison

# Pairwise Tukey HSD
mc = MultiComparison(df['Comfort'], df['Material'])
tukey = mc.tukeyhsd(alpha=0.05)
print(tukey.summary())


#Compact Letter Display (which groups are statistically the same)
# This gives us group letters like A/B; groups sharing a letter are NOT significantly different

def compact_letter_display(tukey_result):
    groups = mc.groupsunique
    reject_table = pd.DataFrame(tukey_result._results_table.data[1:],
                                columns=tukey_result._results_table.data[0])
    # build adjacency of "no difference"
    k = len(groups)
    same = {g:set([g]) for g in groups}
    for _, row in reject_table.iterrows():
        a, b, reject = row['group1'], row['group2'], row['reject']
        if not reject:
            same[a].add(b)
            same[b].add(a)

    # greedy letter assignment
    letters = {}
    used_letters = []
    for g in sorted(groups, key=lambda x: df[df['Material']==x]['Comfort'].mean(), reverse=True):
        placed = False
        for L in used_letters:
            # can g join letter L if compatible with all groups already there?
            if all((h in same[g]) for h in letters if letters[h]==L):
                letters[g] = L
                placed = True
                break
        if not placed:
            newL = chr(ord('A')+len(used_letters))
            used_letters.append(newL)
            letters[g] = newL
    return letters

letters = compact_letter_display(tukey)
print("\nCompact Letter Display (same letter = not significantly different):")
for g, L in sorted(letters.items()):
    print(f"{g}: {L}")

#Show means with letters, sorted
means = df.groupby("Material")["Comfort"].mean().sort_values(ascending=False)
mean_letters = pd.DataFrame({"Mean": means})
mean_letters["Group"] = [letters[g] for g in mean_letters.index]
print("\nGroup means with CLD:\n", mean_letters)



#Interpretation
# Silicone had significantly higher comfort than all other materials.
# Rubber and Plastic were each significantly higher than Metal, but not significantly different from each other.
# Metal had the lowest comfort ratings.
# Compact letter display results showed Silicone (A) > Rubber (B) / Plastic (B) > Metal (C).
```

```python
# Question 4 Section 2
# A randomized block design is appropriate because prior experience with elderly patients may influence the outcome (attitude scores).
# By blocking on this variable, we control for its variability and increase the precision of the treatment comparison.
# Why a CRD would not be appropiate:
  # A CRD ignored Prior Experience, so experience-related differences end up in the residual (MSE), inflating error variance and reducing power to detect the method effect.
  # If randomization happens to produce imbalance in experience across methods in your realized sample, CRD can also confound the treatment comparison (apparent method differences could be partly due to experience).
  # Blocking guarantees like-with-like comparisons (VR vs Traditional within each experience level), which CRD does not.


  #Code to justify why CRD would be suboptimal by demonstrating an experience effect

import pandas as pd
from scipy.stats import ttest_ind

nursing = pd.read_excel("Nursing VR Study.xlsx")
# Check the actual column names in the nursing dataset
print(nursing.columns.tolist())

# Rename columns to match what we've been using in the analysis
nursing = nursing.rename(columns={
    'Method': 'Teaching_Method',
    'Experience': 'Prior_Experience',
    'Score': 'Attitude_Score'
})

print(nursing.columns)  # should now show the new names

from scipy.stats import ttest_ind
import pandas as pd

# Check cell counts
print("Cell counts:")
print(pd.crosstab(nursing['Teaching_Method'], nursing['Prior_Experience']))

# Look at experience effect collapsed across teaching methods
exp_groups = nursing.groupby('Prior_Experience')['Attitude_Score']
print("\nMeans by Prior_Experience:")
print(exp_groups.mean())
print("\nSDs by Prior_Experience:")
print(exp_groups.std())

# t-test for quick experience effect check
a = nursing.loc[nursing['Prior_Experience'] == 'Yes', 'Attitude_Score']
b = nursing.loc[nursing['Prior_Experience'] == 'No',  'Attitude_Score']
t_stat, p_val = ttest_ind(a, b, equal_var=True)
print(f"\nExperience effect (pooled): t={t_stat:.3f}, p={p_val:.4f}")

if p_val < 0.05:
    print("=> Prior experience significantly affects scores. RBD justified.")
else:
    print("=> Blocking still prudent even if not significant, to control variance.")
```

```python
#Question 5 Section 2

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics.factorplots import interaction_plot

# Load & clean
nursing = pd.read_excel("Nursing VR Study.xlsx").rename(columns={
    'Method': 'Teaching_Method',
    'Experience': 'Prior_Experience',
    'Score': 'Attitude_Score'
})

# Drop missing rows if any
nursing = nursing[['Teaching_Method', 'Prior_Experience', 'Attitude_Score']].dropna()
nursing['Teaching_Method'] = nursing['Teaching_Method'].astype(str) # Convert to string
nursing['Prior_Experience'] = nursing['Prior_Experience'].astype(str) # Convert to string


# Summary stats
by_method = nursing.groupby('Teaching_Method')['Attitude_Score'].describe()
by_exp = nursing.groupby('Prior_Experience')['Attitude_Score'].describe()
by_cell = nursing.groupby(['Teaching_Method','Prior_Experience'])['Attitude_Score'].describe()

print("\nSummary by Teaching Method:\n", by_method)
print("\nSummary by Prior Experience:\n", by_exp)
print("\nSummary by Method x Experience:\n", by_cell)

# Boxplot: Treatment effect by block
plt.figure(figsize=(8,5))
sns.boxplot(data=nursing, x='Teaching_Method', y='Attitude_Score', hue='Prior_Experience')
plt.title('Attitude Scores by Teaching Method and Prior Experience')
plt.show()

# Interaction plot
plt.figure(figsize=(8,5))
interaction_plot(
    x=nursing['Teaching_Method'],
    trace=nursing['Prior_Experience'],
    response=nursing['Attitude_Score'],
    colors=['blue', 'red']
)
plt.title('Interaction: Teaching Method √ó Prior Experience')
plt.ylabel('Mean Attitude Score')
plt.show()


#Interpretation

#The descriptive statistics show that students in the VR group had a notably higher mean attitude score (M = 40.75, SD = 6.90) than those in the Traditional group (M = 31.30, SD = 5.68).
# Students with prior experience (M = 36.15) and those without prior experience (M = 35.90) had similar overall scores, suggesting a weak block effect.
# The boxplots illustrate clear separation between VR and Traditional groups, with VR consistently higher across both experience levels.
# The interaction plot shows approximately parallel lines, indicating little to no interaction between Teaching Method and Prior Experience.
# These exploratory findings support the alternative hypothesis that teaching method affects attitude scores.‚Äù
```

```python
#Question 7 Section 2
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import shapiro
import seaborn as sns
import matplotlib.pyplot as plt

# Two-way ANOVA model
model_2way = ols('Attitude_Score ~ C(Teaching_Method) + C(Prior_Experience)', data=nursing).fit()

# Shapiro‚ÄìWilk test on residuals
residuals = model_2way.resid
shapiro_stat, shapiro_p = shapiro(residuals)
print(f"Shapiro-Wilk Test: Statistic={shapiro_stat:.4f}, p={shapiro_p:.4f}")

# Q-Q plot
sm.qqplot(residuals, line='s')
plt.title("Q-Q Plot of Residuals (Two-Way ANOVA)")
plt.show()

# Histogram of residuals
sns.histplot(residuals, kde=True)
plt.title("Histogram of Residuals (Two-Way ANOVA)")
plt.show()

#Interpretation
# The Shapiro‚ÄìWilk test returned a p-value of 0.3495, which is greater than the 0.05 significance level. Therefore, we fail to reject the null hypothesis of normality.
# The Q‚ÄìQ plot and histogram show the residuals are approximately normally distributed, supporting the normality assumption of the two-way ANOVA model.
```

```python
#Question 8 Section 2

from scipy.stats import levene
import matplotlib.pyplot as plt

# Levene's test across the 4 groups (Method x Experience)
groups = [g['Attitude_Score'].values
          for _, g in nursing.groupby(['Teaching_Method','Prior_Experience'])]

lev_stat, lev_p = levene(*groups)
print(f"Levene's Test: Statistic={lev_stat:.4f}, p-value={lev_p:.4f}")

# Residual vs fitted plot
fitted = model_2way.fittedvalues
residuals = model_2way.resid

plt.scatter(fitted, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values (Two-Way ANOVA)')
plt.show()


#Interpretation

#Levene‚Äôs test yielded a p-value of 0.7634, which is greater than 0.05. Therefore, we fail to reject the null hypothesis of equal variances across groups.
# Additionally, the residual vs. fitted plot does not show any noticeable pattern or heteroscedasticity. These results support the assumption of homogeneity of variance for the two-way ANOVA model
```

```python
#Question 9 Section 2
from statsmodels.stats.anova import anova_lm

anova_table_2way = anova_lm(model_2way, typ=2)
print(anova_table_2way)


#Interpretation

# The two-way ANOVA revealed a significant main effect of Teaching Method on attitude scores, F(1, 37) = 21.81, p < 0.001, indicating that students in the VR condition scored significantly higher than those in the Traditional condition.
#There was no significant effect of Prior Experience, F(1, 37) = 0.015, p = 0.902, suggesting that prior experience with elderly patients did not influence attitude scores in this sample. Since the block variable was not significant, it did not account for much variability, but including it still helps ensure control for potential confounding.
#These results strongly support rejecting the null hypothesis for Teaching Method in favor of the alternative.
```

```python
#Question 1 Section 3

# Because humans vary a lot on important characteristics (e.g., prior ability, age, baseline attitudes), a one-factor completely randomized design (CRD) tends to:
 # Leave major sources of variability in the error term
 # Risk imbalance/confounding if those traits aren't evenly split by chance
 # Ignore ethical/practical needs to ensure comparability
 # Be more vulnerable to noncompliance/attrition effects that differ by subgroup
 # Hence, the designs that control person-level heterogeneity (RBD, matched pairs, crossover) are usually preferred.
```

```python
# Question 2 Section 3
# The purpose of post-hoc tests are used after a statistically significant overall ANOVA to determine which specific group means differ from each other while controlling the overall Type I error rate.
# ANOVA only tells us that at least one group differs.
# Post-hoc tests identify where those differences occur.
# They adjust for multiple comparisons, ensuring that the probability of making a false positive error across all pairwise tests stays at or below the chosen significance level
```

```python
#Question 3 Section 3
#RBD:
 # Participants or experimental units are grouped into ‚Äúblocks‚Äù based on a factor that is expected to influence the outcome (e.g., prior experience, age group, skill level).
 # Within each block, treatments are randomly assigned
 # The goal is to remove variability due to the blocking factor from the error term
 # Useful when you know one major nuisance variable you want to control

 #LSD:
 # Used when there are two blocking factors (e.g., day and time, or gender and location) in addition to the treatment.
 # Each treatment appears exactly once in each row and column, balancing across both blocking factors
 # This design controls for two sources of nuisance variability without requiring a full factorial design
 # Efficient design when multiple nusiance factors must be controlled

 #Key takeway: LSD is essentially an extension of RBD to two blocking factors

```

```python
#Question 4 Section 3

# Randomization is a core principle of experimental design because it helps ensure that treatment groups are comparable and that uncontrolled or unknown factors do not systematically bias the results.
# It reduces bias: Random assignment spreads both known and unknown nuisance variables evenly across groups, minimizing confounding.
# Suppoprts valid statistical inference: It justifies the use of probability theory in hypothesis testing, making p-values and confidence intervals meaningful.
# Improves internal validity: It strengthens the causal interpretation of treatment effects by ensuring group differences are due to the treatment, not pre-existing differences.
# Simplifies analysis: With proper randomization, residual error is ‚Äúnoise,‚Äù not hidden structure, making model assumptions easier to meet.

```

```python
#Question 5 Section 3

# Replication means repeating treatments on multiple experimental units (e.g., participants, test subjects, or trials) rather than relying on a single observation per treatment.
 # It is one of the three fundamental principles of experimental design (along with randomization and blocking).

# It increases precision: Replication reduces the impact of random variation (noise) and gives more stable estimates of treatment effects.
# Allows estimation of experimental error: With multiple observations per group, we can calculate residual variance and conduct valid significance tests.
# Improves power: More replications mean greater ability to detect real differences between treatments.
# Enhances generalizability: Results are less likely to be due to anomalies in a small or unrepresentative sample.
# Supports reproducibility: Replication strengthens the credibility of findings and allows others to verify results.
```
